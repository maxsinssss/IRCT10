```python
# Define the documents
document1 = "The computer science students are appearing for practical examination."
document2 = "computer science practical examination will start tomorrow."

documents = {
    1: document1,
    2: document2
}

import re

# Preprocessing function
def preprocess(text):
    text = text.lower()                  # Lowercase
    text = re.sub(r'[^\w\s]', '', text)   # Remove punctuation
    tokens = text.split()                 # Tokenize
    return tokens

# Preprocess all documents
preprocessed_documents = {}
for doc_id, content in documents.items():
    preprocessed_documents[doc_id] = preprocess(content)

print(preprocessed_documents)

# Build inverted index
inverted_index = {}

for doc_id, tokens in preprocessed_documents.items():
    for token in tokens:
        if token not in inverted_index:
            inverted_index[token] = []
        if doc_id not in inverted_index[token]:
            inverted_index[token].append(doc_id)

# Display Inverted Index
print("Inverted Index:")
for term, doc_list in inverted_index.items():
    print(f"{term} → {doc_list}")

# Retrieve documents containing both 'computer' and 'science'
term1_docs = set(inverted_index.get('computer', []))
term2_docs = set(inverted_index.get('science', []))

# Intersection: documents containing both terms
result_docs = term1_docs.intersection(term2_docs)

print("\nDocuments containing 'computer science':", result_docs)

```
