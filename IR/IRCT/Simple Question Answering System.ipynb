```python
# Simple rule-based Question Answering
def simple_qa_system(question):
    question = question.lower()

    if "computer science students" in question and "doing" in question:
        return "They are appearing for practical examination."

    if "practical examination" in question and "when" in question:
        return "It will start tomorrow."

    return "Sorry, I don't know the answer."

# Testing QA System
question1 = "What are computer science students doing?"
question2 = "When will the practical examination start?"

print("Q:", question1)
print("A:", simple_qa_system(question1))

print("\nQ:", question2)
print("A:", simple_qa_system(question2))
```
