```python

!pip install networkx matplotlib

import networkx as nx
import matplotlib.pyplot as plt


G = nx.DiGraph()


G.add_edges_from([
    ("A", "B"), ("A", "C"), ("A", "D"),
    ("B", "C"), ("B", "E"),
    ("C", "A"), ("C", "D")
])


pagerank_scores = nx.pagerank(G, alpha=0.85)


print("PageRank Scores:")
for node, score in pagerank_scores.items():
    print(f"Page {node}: {score:.4f}")


plt.figure(figsize=(8, 6))
pos = nx.spring_layout(G, seed=42)

node_sizes = [10000 * pagerank_scores[node] for node in G.nodes()]
nx.draw_networkx_nodes(G, pos, node_size=node_sizes, node_color='lightgreen', alpha=0.9)
nx.draw_networkx_edges(G, pos, arrowstyle='->', arrowsize=20, edge_color='gray')
nx.draw_networkx_labels(G, pos, font_size=14, font_weight='bold')

plt.title("Simplified Web Graph with PageRank Scores", fontsize=16)
plt.axis("off")
plt.show()
```
