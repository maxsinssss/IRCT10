```python

documents = {
    1: "BSc lectures start at 7.",
    2: "My lectures are over.",
    3: "Today is a holiday."
}

normalized_docs = {doc_id: doc.lower().split() for doc_id, doc in documents.items()}


query_word = "lectures"


result_docs = [doc_id for doc_id, tokens in normalized_docs.items() if query_word not in tokens]

# Output results
print("Documents that match the query 'NOT lectures':", result_docs)
```
